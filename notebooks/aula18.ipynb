## Ficha 9 - Que robôs podem dar a mão?



```haskell
[4,1,1,2] |> sum |> reverse
```

```haskell
[4,1,1,2] |> reverse |> nub
```

```haskell
[4,1,1,2] |> nub |> reverse
```

```haskell
[4,1,1,2] |> reverse |> length
```

```haskell
[4,1,1,2] |> maximum |> sum
```

```haskell
[4,1,1,2] |> length |> maximum
```
