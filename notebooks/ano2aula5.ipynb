```haskell
[5,2,7,1,8,6,4,9,3,1]
```

Será que o sinal `+` também serve para juntar listas?

```haskell
[5,2,7,1,8,6,4,9] + [3,1]
```

Alguma sugestão? 🤔💡

```haskell
[5,2,7,1,8,6,4,9] ? [3,1]
```

---
E não funciona só com 2 listas de números!
<br>Podemos juntar **mais do que 2 listas**.
<br>E podemos fazer junções de **listas de números**, ou de **listas de caracteres**, ou de **listas de strings**, ou de listas com outros tipos de elementos.

```haskell
[1,2,3] ++ [4,5] ++ [6,7,8,9]
```

```haskell
"eu fui ali" ++ " e ali"
```

```haskell
["batata", "cebola"] ++ ["couve", "cenoura"]
```

Desde que as listas tenham **elementos do mesmo tipo**!

```haskell
["isto", "são", "strings"] ++ [17,45]
```

Se tentarmos misturar alhos com bugalhos, dá erro! ❌🤪

## Exercícios

```haskell
["Helena", "Manuel"] ++ ["Ricardo", "Sara", "Teresa"]
```

```haskell
["Helena", "Manuel", "Ricardo"] ++ ["Sara", "Teresa"]
```

```haskell
[1,2,3,4] ++ [4,5,6]
```

```haskell
[3,2,1] ++ []
```

```haskell
["Ines", "Luis"] ++ ["Rita", "Liliana"]
```

```haskell
["Rita", "Liliana"] ++ ["Ines", "Luis"]
```

Importa sim! Se juntarmos as listas por ordens diferentes, os resultados **são diferentes**.

```haskell
["Ines", "Luis"] ++ []
```

```haskell
[] ++ ["Ines", "Luis"]
```
